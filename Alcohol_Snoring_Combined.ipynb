```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url='https://drive.google.com/file/d/1IbHzUMYKFBkmW3TIvryP7XJlW-WkmGEw/view?usp=sharing'
path= 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
df = pd.read_csv(path)
df.head()
```

```python
#Manage data for variable 1
df['H4TO33'].value_counts()
```

```python
df['Drink_Alcohol'] = df['H4TO33']
df['Drink_Alcohol'].value_counts()
```

```python
df['Drink_Alcohol'].replace(['0','1','7'] , ['NO','YES','NO'] , inplace = True)
df['Drink_Alcohol'].value_counts() 
```

```python
#Manage data for variable 1
df['H4SP7'].value_counts()
```

```python
df['SNORING'] = df['H4SP7']
df['SNORING'].value_counts()
```

```python
df['SNORING'] = df['SNORING'].loc[ df['SNORING'] !='5'] 
df['SNORING'] = df['SNORING'].loc[ df['SNORING'] !='8'] 
df['SNORING'] = df['SNORING'].loc[ df['SNORING'] !=' '] 
df['SNORING'].value_counts()
```

```python
#Graphs and tables or statistics
rp.summary_cat(df['SNORING'])
```

```python
explode = (0.1, 0.01)
df['SNORING'].value_counts().plot(kind ='pie', autopct='%1.1f%%',  explode=explode, shadow=True, startangle=90, colors=['blue', 'yellow'])
plt.legend(loc ="lower right")
```

```python
plt.xlabel("Amount of Limitations")
plt.ylabel("Percent")
plt.title("Based on what you have noticed or what others have told you, are there times when you snore or you stop breathing during your sleep?")
df['SNORING'].value_counts(normalize = True).plot(kind ='barh', color = 'red', width = 0.2, alpha = 0.5, grid = True)
```
